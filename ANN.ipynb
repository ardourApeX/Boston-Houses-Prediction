````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

``` Load Dataset```

````
df = pd.read_csv("./dataset/Train_Data.csv")
df
````

## Data Cleaning

````
def is_there_null(df):
  null_values = list(df.isnull().sum())
  columns = df.columns
  null_values = {columns[i] : null_values[i] for i in range(len(columns)) if null_values[i] != 0}
  return null_values
````

````
null_values = is_there_null(df)
print(null_values)
````

```Droping those columns which contain maximum of null values```

````
columns_to_drop = ["Id", "Alley", "Fence", "FireplaceQu", "MiscFeature", "PoolQC"]
df.drop(columns = columns_to_drop, inplace = True)
df.head()
````

````
null_values = is_there_null(df)
print(null_values)
````

## Filling NaN values and Label Encoding

````
for column in df.columns:
  if df[column].dtype == "object":
    df[column] = df[column].astype("category")
    df[column] = df[column].cat.codes  
  elif df[column].isnull().any():
    df[column].fillna(value = round(df[column].mean()), inplace = True)
    del null_values[column]
````

````
print(null_values)
````

## Label Encoders make Nan values equal to -1 So it is better to replace them with mean values

````
df.head()
````

````
for key in null_values.keys():
  df.loc[df[key] == -1, key] = round(df[key].mean())
````
