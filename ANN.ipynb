````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

``` Load Dataset```

````python
df = pd.read_csv("/content/Train_Data.csv")
df.head()
````

## Data Cleaning

````python
def is_there_null(df):
    '''Return those columns which are having any null values along with frequency in a dictionary format'''
    null_values = list(df.isnull().sum())
    columns = df.columns
    null_values = {columns[i] : null_values[i] for i in range(len(columns)) if null_values[i] != 0}
    return null_values
````

````python
null_values = is_there_null(df)
print(null_values)
````

```Droping those columns which contain maximum of null values```

````python
columns_to_drop = ["Id", "Alley", "Fence", "FireplaceQu", "MiscFeature", "PoolQC"] # These columns are having maximum no of null values
df.drop(columns = columns_to_drop, inplace = True)
df.head()
````

````python
null_values = is_there_null(df) # Now check how many columns have null values
print(null_values)
````

## Filling NaN values and Label Encoding

````python
for column in df.columns:
    '''This function either converts the dtype from object to category for label encoding or fill the null values of other dtype'''
  
    if df[column].dtype == "object":
        df[column] = df[column].astype("category") #Converting dype
        df[column] = df[column].cat.codes  # Label Encoding
    elif df[column].isnull().any(): # if any null value exist
        df[column].fillna(value = round(df[column].mean()), inplace = True) #fill it with mean value
        del null_values[column] #And remove it from record
````

````python
print(null_values) # Now we have only those null values upon which we have applied label encoding
````

```Label Encoders make Nan values equal to -1 So it is better to replace them with mean values```

````python
df.head()
````

````python
for key in null_values.keys():
    df.loc[df[key] == -1, key] = round(df[key].mean()) #If any -1 value is encountered make it equal to mean
````

````python
# Cross Check
(df["MasVnrType"] == -1).any()
````

## Feature Selection

````python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

# Separating X and Y from Data frame
X = df.iloc[:,:-1]
Y = df.iloc[:, -1]

best_features = SelectKBest(score_func = chi2, k = 20) # Object Instantiation

#finding Relevancy of each feature
best_fit = best_features.fit(X, Y)
best_fit.scores_
````

````python
dfscores = pd.DataFrame(best_fit.scores_)
````

````python
dfbest = pd.DataFrame(X.columns)
dfbest
````

````python
# Now concatenate them with scores

feature_score = pd.concat([dfbest, dfscores], axis = 1)

feature_score.columns = ["feature", "score"]
````

````python
feature_score.head()
````

````python
# Now bring up the more relevant features

feature_score.sort_values(by = "score", ascending = False, inplace = True)
````

````python
feature_score.reset_index(drop = True, inplace = True)
````

## Visualization

## Applying ANN 

````python
from keras.models import Sequential
from keras.layers import Dense
````

````python
classifier = Sequential()
````

Adding Layers

````python
classifier.add(Dense(100, input_dim = 30, kernel_initializer = "normal", activation = "relu"))
classifier.add(Dense(50, kernel_initializer = "normal", activation = "relu"))
classifier.add(Dense(1, kernel_initializer = "normal", activation = "linear"))
classifier.compile(loss = "mean_squared_error", optimizer = 'adam', metrics = ["accuracy"])
````

````python
X = X.values
Y = Y.values
````

````python
trained_model = classifier.fit(X, Y, epochs = 1000, batch_size = 30)
````

````python
plt.plot(trained_model.history["loss"], color = "green", )
````
