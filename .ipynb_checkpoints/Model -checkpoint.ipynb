## Normalization

It refers to the scaling down of the data set such that the normalized data falls in the range between 0 and 1

$X_{i}^{Normalized} = \frac{X_{i} - \overline{X}}{\sigma}$


```python
mean = X.mean()
sigma = X.std()

X = (X - mean)/sigma
```

```python
# Scatter Plot

plt.scatter(X, Y, color = 'blue')
plt.xlabel("Hardwork")
plt.ylabel("Performance")
plt.title("Hardwork vs Performance");
```

## Hypothesis Function

$ h(\theta) = \theta_{0} + \theta_{1}X$

#### Through out the Model 
$Y\_predict = h(\theta)$

```python
def Hypothesis(X, theta):
    hypo = theta[0] + theta[1] * X
    return hypo
    
```

## Gradient

$ Cost Function = J(\theta_{0}, \theta_{1}) = \frac{1}{2m} \sum_{i=1}^{m}( h(\theta) - Y)^2 $

$ Gradient = \frac{\partial J(\theta_{0}, \theta_{1})}{\partial \theta(_j)} $

$ Gradient[0] = \frac{1}{m} \sum_{i=1}^{m}( h(\theta) - Y)$\
$ Gradient[1] = \frac{1}{m} \sum_{i=1}^{m}( h(\theta) - Y)X_{i}$

```python
def Gradient(X, Y, theta):
    '''This Function is to calculate derivative of Cost Function with respect to theta0 & theta1'''
    
    m = X.shape[0]  # Number of examples
    gradient = np.zeros((2, 1)) # Initializing Gradient
    
    for i in range(m):
        #ith example
        x = X[i] 
        y = Y[i]
        
        Y_predict = Hypothesis(x, theta)
        
        gradient[0] += (Y_predict - y)
        gradient[1] += (Y_predict - y)*x
        
    return gradient/m
```

## Error

$ Error = \frac{1}{m} \sum_{i=1}^{m}( h(\theta) - Y)^2$

$ Error = Cost Function = \frac{1}{m} \sum_{i=1}^{m}( h(\theta) - Y)^2 $

#### Note : 
$\frac{1}{2}$ is of no use as it can be manged by learning rate ie ...\
$\frac{1}{2}$ * learning_rate' = learning_rate (where learning_rate = $\frac{learning\_rate'}{2})$


```python
def error(X, Y, theta):
    ''' This function is to find TOTAL ERROR for a particular Theta over whole Dataset'''
    
    m = X.shape[0]    
    Error = 0.0
    
    for i in range(m):
        
        Y_predict = Hypothesis(X[i], theta) 
        Error += (Y_predict - Y[i])**2 # Total error for a particular set of theta over whole dataset
    
    return Error / m
```

## Gradient Descent

$\theta_{0} = \theta_{0} - \alpha \frac{\partial J(\theta_{0}, \theta_{1})}{\partial \theta(_0)}$

$\theta_{1} = \theta_{1} - \alpha \frac{\partial J(\theta_{0}, \theta_{1})}{\partial \theta(_1)}$

```python
theta = np.zeros((2,1))
```

```python
def GradientDescent(X, Y, LearningRate = 0.1, iteration = X.shape[0]):
    '''This Function is to minimize the error and cost by converging theta to global minima'''
    
    theta = np.zeros((2,)) # Initialized with 0
    error_list = [] # store average error for each change in theta over whole dataset
    theta_list = [] # store history of changes occured in theta
   
    for i in range(iteration):
        
        gradient = Gradient(X, Y, theta) # Derivative 
        error_list.append(error(X, Y, theta))
        
        # Updation in theta
        theta[0] = theta[0] - LearningRate * gradient[0]
        theta[1] = theta[1] - LearningRate * gradient[1]
        
        theta_list.append((theta[0], theta[1]))
      
        # Can be placed before updation in theta as well
        
    return theta, error_list, theta_list

        
```

### When Number of iterations are extremely large
> It may cause overfitting

```python

Theta, Error_list, Theta_list = GradientDescent(X, Y, LearningRate = 0.1, iteration = X.shape[0])

# NumPy array
Theta_list = np.array(Theta_list)

# Plot
plt.plot(Theta_list[:, 0], label = 'Theta0')
plt.plot(Theta_list[:, 1], label = 'Theta1')
plt.title("History of theta")
plt.legend();
```

### When Number of iterations are comparitively smaller

```python
Theta, Error_list, Theta_list = GradientDescent(X, Y, LearningRate = 0.1, iteration = 500 )
```

## $R^2 Score$

$R^2 Score = 1 -  \sum_{i=1}^{m} \frac{(Y{^i} - Y\_predicted)^2}{(Y^i - \overline{Y})^2}$

```python
def R2Score(Y_Predict, Y):
    Numerator = np.sum((Y - Y_Predict) ** 2)
    Denominator = np.sum((Y - np.average(Y)) ** 2)
    score = 1 - Numerator / Denominator
    return score
```

```python
R2Score(Y_predict, Y)
```

```python
# We will create some of the demo Thetas to create a bowl shaped plot

T0 = np.arange(-50, 50, 1) # Theta[0] lies in btw
T1 = np.arange(40, 120, 1) # Theta[1] lies in btw
```

```python
# Creating all possible coordinates

T0, T1 = np.meshgrid(T0, T1)

T0.shape, T1.shape
```

```python
J = np.zeros((T0.shape)) 
J.shape
```

```python
# We will be using Vectorization throughout

# Loop for 80x100 iterations
for i in range(J.shape[0]):
    for j in range(J.shape[1]):
        
        # Hypothesis : Theta0 = T0, Theta1 = T1, Data = X
        Y_Output = T0[i,j] + T1[i,j] * X    # Vectorization
        
        # Cost Function : Squared Mean Error
        J[i,j] = np.sum((Y - Y_Output) ** 2)/X.shape[0]
```

```python
J.shape # Matches to that of meshgrid(T0, T1)
```
