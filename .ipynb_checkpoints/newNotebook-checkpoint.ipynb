### ```Observation No.4```
```New Features```

if YearBuilt  < YearRemodAdd = IsRemod

YrSold  - YearBuilt =   YearsOld

YrSold  - YearRemodAdd = YearsRemod

YrSold - GarageYrBlt = YearsOldGarage

````python
def Feature_Engineering(df):
    
    
#     Remove columns where the variability id very low( if a single class appears for more than 90% of rows)
#     Utilities
#     CentralAir
#     Street
#     BsmtHalfBath
#     LandSlope
#     PavedDrive
#     BsmtCond
#     KitchenAbvGr 
#     Electrical 
#     GarageQual
#     GarageCond
#     Heating
#     RoofMatl
#     Condition2
#     PoolArea 
    
#     Columns need to be deleted as they either comprise of null values or are skewed
    cols_to_del = ['Id','Alley','PoolQC','Fence','MiscFeature','Utilities','CentralAir','Street','BsmtHalfBath','LandSlope','PavedDrive','BsmtCond','KitchenAbvGr','Electrical','GarageQual','GarageCond','Heating','RoofMatl', 'Condition2','PoolArea']
    
    df = df.drop(cols_to_del, axis = 1)
    
#     Impute the null values from categorical column with constants
    fill_dict = {('MasVnrType',): 'None',
                 ('MasVnrArea','GarageYrBlt','LotFrontage') : 0,
                 ('BsmtQual','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType','GarageFinish') : "NA"}
    
    for keys,values in fill_dict.items():
        for cols in keys:
            df[cols].fillna(values,inplace = True)
            
#     Other values for imputation of other columns
    impute_median = ['BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','GarageArea']
    impute_mode = ['Exterior1st','Exterior2nd','BsmtFullBath','SaleType','GarageCars','MSZoning','KitchenQual', 'Functional']
   
    for cols in impute_median :
        df[cols] = df[cols].fillna(df[cols].median())
        
    for cols in impute_mode :
        df[cols] = df[cols].fillna(df[cols].mode()[0])
    
#     Make Categorical Columns 
    cat_cols_1 = ['ExterQual', 'ExterCond', 'BsmtQual', 'HeatingQC', 'KitchenQual','FireplaceQu']
    cat_1 = ['Ex','Gd','TA','Fa','Po','NA']
    cat_1_mapping = {'Ex':6,'Gd':5,'TA':4,'Fa':3,'Po':2,'NA':1}

    cat_cols_2 = ['BsmtFinType1', 'BsmtFinType2']
    cat_2 = ['GLQ','ALQ','BLQ','Rec','LwQ','Unf','NA']
    cat_2_mapping = {'GLQ':7,'ALQ':6,'BLQ':5,'Rec':4,'LwQ':3,'Unf':2,'NA':1}


    cat_cols_3 = ['BsmtExposure']
    cat_3 = ['Gd','Av','Mn','No','NA']
    cat_3_mapping = {'Gd':5,'Av':4,'Mn':3,'No':2,'NA':1}
    

    for cols in cat_cols_1 :
        df[cols] = df[cols].map(cat_1_mapping)
    for cols in cat_cols_2 :
        df[cols] = df[cols].map(cat_2_mapping)
    for cols in cat_cols_3 :
        df[cols] = df[cols].map(cat_3_mapping)
    
# Creating New Features based on previous ones
    df['YearsOld'] = df['YrSold'] - df['YearBuilt']
    df['IsRemod'] = (df['YearBuilt'] < df['YearRemodAdd']).astype(int)
    df['YearsRemod'] = df['YrSold'] - df['YearRemodAdd']
    df['YearsOldGarage'] = df['YrSold']-df['GarageYrBlt']


    df[df['YearsOld'] < 0]['YearsOld'] = 0
    df[df['YearsRemod'] < 0]['YearsRemod'] = 0
    df[df['YearsOldGarage'] < 0]['YearsOldGarage'] = 0
    
    cols_to_del = ['YrSold','YearBuilt','MoSold','YearRemodAdd','GarageYrBlt']
    df = df.drop(cols_to_del,axis = 1)
    
    return df
````

````python
x_train = Feature_Engineering(data)
````
