# Training Model : Vectorization

# Hypothesis Function

$ h(\theta) = \begin{bmatrix}
           \theta_{0} \\
           \theta_{1} \\
           \theta_{2} \\
           \vdots \\
           \theta_{n}
         \end{bmatrix} . \begin{bmatrix} 
    1 & a_{00} & a_{01} & \dots & \dots & \dots & \dots & \dots & a_{0n}\\
    1 & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
    \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
     \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
    1 & a_{m0} & \dots & \dots & \dots & \dots & \dots & \dots & a_{mn} 
    \end{bmatrix} = \begin{bmatrix}
           H_{0} \\
           H_{1} \\
           H_{2} \\
           \vdots \\
           H_{m}
         \end{bmatrix} 
$            

```python
def Hypothesis(X, theta):
    return np.dot(X, theta)
```

## Error

$ error = \frac{1}{m} \sum_{i=1}^{m} \biggl(\begin{bmatrix}
           H_{0} \\
           H_{1} \\
           H_{2} \\
           \vdots \\
           H_{m}
         \end{bmatrix} - \begin{bmatrix}
           Y_{0} \\
           Y_{1} \\
           Y_{2} \\
           \vdots \\
           Y_{m}
         \end{bmatrix}
         \biggr) ^ {2}$

```python
def error(X, Y , theta):
    
    error = 0.0
    m = X.shape[0]
    
    Y_predict = Hypothesis(X, theta) # Y_Predict
    
    error = np.sum((Y - Y_predict) ** 2)
    return error/m
```


## Gradient 

$  \begin{bmatrix} 
    1 & 1 & 1 & \dots & \dots & \dots & \dots & \dots & 1\\
    a_{01} & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
    \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
     \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
    a_{0n} & a_{1n} & \dots & \dots & \dots & \dots & \dots & \dots & a_{mn} 
    \end{bmatrix} . \biggl(\begin{bmatrix}
           H_{0} \\
           H_{1} \\
           H_{2} \\
           \vdots \\
           H_{m}
         \end{bmatrix} - \begin{bmatrix}
           Y_{0} \\
           Y_{1} \\
           Y_{2} \\
           \vdots \\
           Y_{m}
         \end{bmatrix}
         \biggr)$

```python
def gradient(X, y, theta):
    
    n = X.shape[1]
    m = X.shape[0]
    Y_predict = Hypothesis(X, theta)
    
    grad = np.dot(X.T, (Y_predict - Y))
    return grad/m
```

## Gradient Descent
$
\begin{bmatrix}
           \theta_{0} \\
           \theta_{1} \\
           \theta_{2} \\
           \vdots \\
           \theta_{n}
         \end{bmatrix} := \begin{bmatrix}
           \theta_{0} \\
           \theta_{1} \\
           \theta_{2} \\
           \vdots \\
           \theta_{n}
         \end{bmatrix} - \begin{bmatrix} 
    1 & 1 & 1 & \dots & \dots & \dots & \dots & \dots & 1\\
    a_{01} & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
    \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
     \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots\\
    a_{0n} & a_{1n} & \dots & \dots & \dots & \dots & \dots & \dots & a_{mn} 
    \end{bmatrix} . \biggl(\begin{bmatrix}
           H_{0} \\
           H_{1} \\
           H_{2} \\
           \vdots \\
           H_{m}
         \end{bmatrix} - \begin{bmatrix}
           Y_{0} \\
           Y_{1} \\
           Y_{2} \\
           \vdots \\
           Y_{m}
         \end{bmatrix}
         \biggr)$

```python
def GradientDescent(X, Y, LR = 0.1, maxiter = 300):
    n = X.shape[1]
    error_list = []
    theta = np.zeros((n,))
    
    for i in range(maxiter):
        error_list.append(error(X, Y, theta))        
        theta = theta - gradient(X,Y, theta) * LR
    
    return theta, error_list
        
```

```python
import time

start = time.time()
theta, error_list = GradientDescent(X, Y)
end = time.time()
print(end - start)
```

```python
plt.style.use("seaborn")
plt.plot(error_list)
theta
```

```python
Y_predict = Hypothesis(X, theta)
```

```python

def R2score(Y, Y_prediction):
    Numerator = np.sum((Y - Y_prediction) ** 2)
    Denominator = np.sum((Y - np.mean(Y_prediction)) ** 2)
    score = 1 - (Numerator/Denominator)
    return score
```
